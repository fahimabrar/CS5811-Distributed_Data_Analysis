```python
# Remove Unnamed column from timeFormat and Ingredients datasets
ingredientsCount = ingredientsCount[['RecipeId','IngredientsCount']]
ingredientsCount.head()
```

```python
timeFormat = pd.read_csv('C:/Users/SLL26/Desktop/Data Science and Analytics/CS5706 MACHINE LEARNING/CS5811 Coursework/Cleaned CSV/TimeFormat.csv')
```

```python
timeFormat = timeFormat[['RecipeId','CookTime','PrepTime','TotalTime']]
timeFormat.head()
```

```python
# Rename the column names for timeFormat
timeFormat=timeFormat.rename({'CookTime':'CookTimeMins','PrepTime':'PrepTimeMins','TotalTime':'TotalTimeMins'}, axis='columns')
```

```python
# Merge the datasets with recipes datasets
recipeNew = recipes.merge(timeFormat, how='left', on='RecipeId')
recipeNew
```

```python
recipeNew = recipeNew.merge(ingredientsCount, how = 'left', on = 'RecipeId')
recipeNew
```

```python
recipeReview = recipeNew.merge(minMaxMean, how='left', on='RecipeId') 
recipeReview
```

```python
reviewclean.sort_values('RecipeId')
```

```python
recipeReview = recipeReview.merge(reviewclean, how='left', on='RecipeId') 
recipeReview
```
