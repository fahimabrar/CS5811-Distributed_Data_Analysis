```python
recipes = recipes[['RecipeId', 'Name','Images', 'RecipeCategory',
        'AggregatedRating', 'ReviewCount', 'Calories', 'FatContent',
       'SaturatedFatContent', 'CholesterolContent', 'SodiumContent',
       'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent',
       'RecipeServings']]
```

```python
type(recipes)
```

```python
recipes.head()
```

```python
recipes.shape
```

```python
time_merged_data = recipes.merge(times, on = ["RecipeId"])
# not specifing the merge type always do a inner join
```

```python
time_merged_data.shape
```

```python
review_merged_data = time_merged_data.merge(review, on = ["RecipeId"], how = "left")
```

```python
review_merged_data.shape
```

```python
review_merged_data.head()
```

```python
ingredients_merged_data =  review_merged_data.merge(ingredients, on = ["RecipeId"])
#this one will be an inner join as indgredints.csv made form recipes dataset itself
```

```python
ingredients_merged_data.shape
```

```python
minmax.shape
```

```python
main_dataset = ingredients_merged_data.merge(minmax, on = ["RecipeId"], how = "left")
```

```python
main_dataset.shape

main_dataset = main_dataset.rename(columns = {"min":"minimum_rating", "max":"maximum_rating", "mean":"mean_of_rating"})
```

```python
main_dataset.to_csv("main_data.csv")
```
