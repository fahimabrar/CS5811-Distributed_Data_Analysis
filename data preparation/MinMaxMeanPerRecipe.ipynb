```python
# Import required library
import pandas as pd
```

```python
# Load the dataset
reviews = pd.read_csv("C:/Users/SLL26/Desktop/Data Science and Analytics/CS5706 MACHINE LEARNING/CS5811 Coursework/reviews.csv")
```

```python
reviews.head()
```

```python
# Apply groupby with aggregate to get the min rating per Recipe
minRating = reviews.groupby('RecipeId').agg({'Rating': [ 'min']})
```

```python
minRating.head()
```

```python
# Apply groupby with aggregate to get the max rating per Recipe
maxRating = reviews.groupby('RecipeId').agg({'Rating': ['max']})
```

```python
maxRating.head()
```

```python
# Apply groupby with aggregate to get the mean rating per Recipe
meanRating = reviews.groupby('RecipeId').agg({'Rating': ['mean']})
```
